**Calculate Trend Conditions and create dataframe**

```python
def calculate_trend_conditions(data):
    ema = data.ewm(span=20, adjust=False).mean()

    bullish_reversal = (data > ema) & (data.shift(1) <= ema.shift(1))
    bearish_reversal = (data < ema) & (data.shift(1) >= ema.shift(1))
    bullish_cont = (data > ema) & (data.shift(1) > ema.shift(1))
    bearish_cont = (data < ema) & (data.shift(1) < ema.shift(1))
    choppy = ~(bullish_reversal | bearish_reversal | bullish_cont | bearish_cont | np.isnan(data))

    # Creating a new DataFrame with trend conditions
    trend_df = pd.DataFrame(index=data.index, columns=data.columns)
    trend_df[choppy] = "Choppy"
    trend_df[bullish_reversal] = "Bullish Reversal"
    trend_df[bearish_reversal] = "Bearish Reversal"
    trend_df[bullish_cont] = "Bullish Continuation"
    trend_df[bearish_cont] = "Bearish Continuation"

    return trend_df
```

```python
trend_df = calculate_trend_conditions(data)
trend_df = trend_df.drop(trend_df.index[0:21])
trend_df.tail()
```

**Calculate Squeeze Conditions and create dataframe**

```python
def calculate_squeeze_conditions(data):
    # Parameters for squeeze conditions
    BB_mult = 2
    KC_mult = 2
    length = 20

    # Calculate Bollinger Bands
    BB_basis = data.rolling(window=length).mean()
    dev = BB_mult * data.rolling(window=length).std()
    BB_upper = BB_basis + dev
    BB_lower = BB_basis - dev

    # Calculate Keltner Channels using EMA and ATR
    ema = data.ewm(span=length, adjust=False).mean()
    atr = data.rolling(window=length).apply(lambda x: x.max() - x.min())
    KC_upper = ema + KC_mult * atr
    KC_lower = ema - KC_mult * atr

    # Define squeeze conditions
    squeeze_condition = (data < BB_upper) & (data > BB_lower) & (data < KC_upper) & (data > KC_lower)

    # Create DataFrame to store squeeze conditions
    squeeze_df = pd.DataFrame(index=data.index, columns=data.columns)
    squeeze_df[squeeze_condition] = 'Squeeze'
    squeeze_df[~(squeeze_condition | np.isnan(data))] = 'No Squeeze'

    return squeeze_df
```

```python
squeeze_df = calculate_squeeze_conditions(data)
squeeze_df = squeeze_df.drop(squeeze_df.index[0:21])
squeeze_df.tail()
```

**Calculate Momentum Conditions and create dataframe**

```python
2# Define a function to calculate momentum and momentum conditions
def calculate_momentum_conditions(data):
    # Calculate the rolling mean of returns over the specified length
    momentum = data.rolling(window=20).mean()

    # Calculate momentum conditions
    positive_rising = (momentum > 0) & (momentum > momentum.shift(1))
    positive_falling = (momentum > 0) & (momentum < momentum.shift(1))
    negative_rising = (momentum < 0) & (momentum > momentum.shift(1))
    negative_falling = (momentum < 0) & (momentum < momentum.shift(1))

    # Create a new DataFrame with squeeze conditions
    momentum_df = pd.DataFrame(index=data.index, columns=data.columns)
    momentum_df[positive_rising] = "Positive Rising"
    momentum_df[positive_falling] = "Positive Falling"
    momentum_df[negative_rising] = "Negative Rising"
    momentum_df[negative_falling] = "Negative Falling"

    return momentum_df
```

```python
momentum_df = calculate_momentum_conditions(data)
momentum_df = momentum_df.drop(momentum_df.index[0:21])
momentum_df.tail()
```

**Generate Signals and create dataframe**

```python
def generate_signals(squeeze_df, momentum_df):
    signals_df = pd.DataFrame(index=squeeze_df.index, columns=squeeze_df.columns)

    for column in squeeze_df.columns:
        squeeze_condition = squeeze_df[column]
        momentum_condition = momentum_df[column]
        trend_condition = trend_df[column]

        # Generate signals based on conditions
        signals = []
        for squeeze, momentum, trend in zip(squeeze_condition, momentum_condition, trend_condition):
            if squeeze == "Squeeze" and momentum == "Positive Rising" and trend == "Bullish Continuation":
                signals.append("Strong Buy")
            elif squeeze == "Squeeze" and momentum == "Positive Rising" and trend == "Bullish Reversal":
                signals.append("High Buy")
            elif squeeze == "Squeeze" and momentum == "Positive Falling" and trend == "Bullish Continuation":
                signals.append("Medium Buy")
            elif squeeze == "Squeeze" and momentum == "Positive Falling" and trend == "Bullish Reversal":
                signals.append("Low Buy")
            elif squeeze == "Squeeze" and momentum == "Negative Rising" and trend == "Bearish Continuation":
                signals.append("Strong Sell")
            elif squeeze == "Squeeze" and momentum == "Negative Rising" and trend == "Bearish Reversal":
                signals.append("High Sell")
            elif squeeze == "Squeeze" and momentum == "Negative Falling" and trend == "Bearish Continuation":
                signals.append("Medium Sell")
            elif squeeze == "Squeeze" and momentum == "Negative Falling" and trend == "Bearish Reversal":
                signals.append("Low Sell")
            elif pd.isna(squeeze) or pd.isna(momentum) or pd.isna(trend):
                signals.append(np.nan)
            else:
                signals.append("Hold")

        signals_df[column] = signals

    return signals_df
```

```python
signals_df = generate_signals(squeeze_df, momentum_df)
signals_df.tail()
```

**Create a weights list**

```python
weights_list = []

# Loop over the rows of signals_df
for index, row in signals_df.iterrows():
    # Count non-NaN values in the row
    weights = row.notna().sum()
    # Append the count to the list
    weights_list.append(weights)

print("Number of stocks: ", weights_list)
```

**Calculate Portfolio Weights and create dataframe**

```python
def create_portfolio_weights(signals_df):
    portfolio_weights = pd.DataFrame(index=signals_df.index, columns=signals_df.columns)

    for i in range(1, len(signals_df)):
        current_signals = signals_df.iloc[i]
        new_portfolio = portfolio_weights.iloc[i-1].copy()

        # Rebalance the portfolio based on signals
        for security, signal in current_signals.items():
            if signal == 'Strong Buy':
                new_portfolio[security] = 10 / weights_list[i]
            elif signal == 'Strong Sell':
                new_portfolio[security] = -10 / weights_list[i]
            elif signal == 'High Buy':
                new_portfolio[security] = 7 / weights_list[i]
            elif signal == 'High Sell':
                new_portfolio[security] = -7 / weights_list[i]
            elif signal == 'Medium Buy':
                new_portfolio[security] = 4 / weights_list[i]
            elif signal == 'Medium Sell':
                new_portfolio[security] = -4 / weights_list[i]
            elif signal == 'Low Buy':
                new_portfolio[security] = 1 / weights_list[i]
            elif signal == 'Low Sell':
                new_portfolio[security] = -1 / weights_list[i]
            elif signal == 'Hold':
                # Access previous security value
                prev_security_value = new_portfolio[security]
                if prev_security_value > 0:
                    new_portfolio[security] = (prev_security_value - 1) / weights_list[i]
                elif prev_security_value < 0:
                    new_portfolio[security] = (prev_security_value + 1) / weights_list[i]
                else:
                    new_portfolio[security] = 0
            elif pd.isna(signal):
                new_portfolio[security] = np.nan
        portfolio_weights.iloc[i] = new_portfolio

    return portfolio_weights
```

```python
portfolio_weights = create_portfolio_weights(signals_df)
portfolio_weights.tail()
```

**Fetch stock prices again**

```python
stock_prices = prices_df.resample('M').last()
stock_prices.tail()
```

**Create Portfolio Investment Dataframe**

```python
def create_portfolio_investment(portfolio_weights, stock_prices):
    portfolio_investment = portfolio_weights.copy()

    for security in portfolio_weights.columns:
        portfolio_investment[security] *= stock_prices[security]

    return portfolio_investment
```

```python
portfolio_investment = create_portfolio_investment(portfolio_weights, stock_prices)
portfolio_investment.tail()
```

**Create Portfolio Value Dataframe**

```python
def create_portfolio_value(portfolio_weights, stock_prices):
    portfolio_value = portfolio_weights.copy()

    for security in portfolio_weights.columns:
        next_month_prices = stock_prices[security].shift(-1)
        portfolio_value[security] *= next_month_prices

    return portfolio_value
```

```python
portfolio_value = create_portfolio_value(portfolio_weights, stock_prices)
portfolio_value.tail()
```

**Create Portfolio Returns Dataframe**

```python
def create_portfolio_returns(portfolio_value, portfolio_investment):
    portfolio_returns = portfolio_weights.copy()

    for security in portfolio_weights.columns:
        portfolio_returns[security] = portfolio_value[security] - portfolio_investment[security]

    return portfolio_returns
```

```python
portfolio_returns = create_portfolio_returns(portfolio_value, portfolio_investment)
portfolio_returns.tail()
```

**Calculate the Initial NAV**

```python
portfolio_investment_ = portfolio_investment.copy().abs()
#portfolio_investment_.fillna(0, inplace=True)
portfolio_investment_['NAV'] = portfolio_investment_.sum(axis=1)
portfolio_investment_.tail()
```

**Calculate the Final NAV**

```python
portfolio_value_ = portfolio_value.copy().abs()
#portfolio_value_.fillna(0, inplace=True)
portfolio_value_['NAV'] = portfolio_value_.sum(axis=1)
portfolio_value_.tail()
```

**Create an analysis dataframe**

```python
analysis = pd.DataFrame()
analysis['Initial NAV'] = portfolio_investment_['NAV']
analysis['Final NAV'] = portfolio_value_['NAV']
analysis = analysis[(analysis != 0).all(axis=1)]
#analysis.drop(analysis.index[0:144], inplace=True)
analysis.tail()
```

**Calculate shares, investment and value**

```python
shares = []
investment = []
value = []

initial_nav = analysis['Initial NAV'].values.tolist()
final_nav = analysis['Final NAV'].values.tolist()

# Assign initial values
shares.append(100)
investment.append(initial_nav[0] * shares[0])
value.append(final_nav[0] * shares[0])

print("Initial values:")
print("Shares:", shares)
print("Investment:", investment)
print("Value:", value)

for i in range(1, len(initial_nav)):
    investment.append(value[i-1])
    shares.append(investment[i] / initial_nav[i])
    value.append(final_nav[i] * shares[i])

print("Final values after iteration:")
print("Shares:", shares)
print("Investment:", investment)
print("Value:", value)
```

**Calculate returns**

```python
returns = [x - y for x, y in zip(value, investment)]
print("Returns:", returns)
```

**Add these to the dataframe**

```python
analysis['Investment'] = investment
analysis['Value'] = value
analysis['Shares'] = shares
analysis['Returns'] = returns
analysis.tail()
```

**Create the ref dataframe which is the MSCI ACWI Index**

```python
ref = pd.DataFrame()

start_date = '1994-03-01'
end_date = '2024-03-01'

data = yf.download('ACWI', start=start_date, end=end_date)
ref['Price'] = data['Adj Close']
ref = ref.resample('M').last()
ref['PnL'] = ref['Price'].diff()
ref['PnL'] = ref['PnL'].shift(-1)
ref = ref.dropna()
ref.tail()
```

*(share price is value divided by 1000 so that the price is attractive)*

```python
country_etf = pd.DataFrame()
country_etf['Price'] = analysis['Value'] / 1000
country_etf['PnL'] = analysis['Returns'] / 1000
country_etf.tail()
```

**Calculate Performance Metrics for Country ETF**

```python
country_etf['Returns'] = country_etf['PnL'] / country_etf['Price']
total_returns_country_etf = (country_etf['Price'].iloc[-1] - country_etf['Price'].iloc[0]) / country_etf['Price'].iloc[0]
annualized_returns_country_etf = (1 + total_returns_country_etf)**(12 / len(country_etf)) - 1
volatility_country_etf = country_etf['Returns'].std()
sharpe_ratio_country_etf = (annualized_returns_country_etf - 0.035) / volatility_country_etf
cumulative_returns_country_etf = (1 + country_etf['Returns']).cumprod()
peak_country_etf = cumulative_returns_country_etf.cummax()
drawdown_country_etf = (cumulative_returns_country_etf - peak_country_etf) / peak_country_etf
max_drawdown_country_etf = drawdown_country_etf.min()
```

**Calculate Performance Metrics for ACWI**

```python
ref['Returns'] = ref['PnL'] / ref['Price']
total_returns_ref = (ref['Price'].iloc[-1] - ref['Price'].iloc[0]) / ref['Price'].iloc[0]
annualized_returns_ref = (1 + total_returns_ref)**(12 / len(ref)) - 1
volatility_ref = ref['Returns'].std()
sharpe_ratio_ref = (annualized_returns_ref - 0.035)/ volatility_ref
cumulative_returns_ref = (1 + ref['Returns']).cumprod()
peak_ref = cumulative_returns_ref.cummax()
drawdown_ref = (cumulative_returns_ref - peak_ref) / peak_ref
max_drawdown_ref = drawdown_ref.min()
```

**Tabulate Performance Metrics**

```python
from prettytable import PrettyTable

table = PrettyTable()
table.field_names = ["Performance Metric", "Country ETF", "ACWI"]
separator_line = ["-" * len(col) for col in table.field_names]

table.add_row(["Total Returns", round(total_returns_country_etf * 100,2), round(total_returns_ref * 100, 2)])
table.add_row(separator_line)
table.add_row(["Annualized Returns", round(annualized_returns_country_etf * 100,2), round(annualized_returns_ref * 100,2)])
table.add_row(separator_line)
table.add_row(["Volatility", round(volatility_country_etf * 100,2), round(volatility_ref * 100, 2)])
table.add_row(separator_line)
table.add_row(["Sharpe Ratio", round(sharpe_ratio_country_etf,2), round(sharpe_ratio_ref, 2)])
table.add_row(separator_line)
table.add_row(["Max Drawdown", round(max_drawdown_country_etf * 100,2), round(max_drawdown_ref * 100, 2)])

print(table)
```
